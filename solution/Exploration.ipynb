```python
# Change the commas to dots
comma_to_dot = ['distance', 'consume', 'temp_inside', 'refill liters']
for col in comma_to_dot:
    measures[col] = measures[col].str.replace(',','.').astype(float)
measures
```

```python
# Fill the null spaces
measures['temp_inside'].fillna(measures['temp_inside'].mean(), inplace=True)
measures['specials'].fillna('None', inplace=True)
measures['refill gas'].fillna('None', inplace=True)
measures['refill liters'].fillna(0, inplace=True)

# Change the names of the columns
measures.columns = measures.columns.str.lower()
measures
```

```python
# Create the dummy
measures = pd.get_dummies(measures, columns=['specials', 'gas_type', 'refill gas'])
```

```python
# Add two new columns
measures['fuel_efficiency'] = measures['distance'] / measures['consume']
measures['temp_difference'] = measures['temp_outside'] - measures['temp_inside']
```

```python
# Drop the outliers
measures = measures[(measures['distance'] > 0) & (measures['consume'] > 0)]
```

```python
measures
```

```python
# Normalize
numerical_cols = ['distance', 'consume', 'speed', 'temp_inside', 'temp_outside', 'refill liters', 'fuel_efficiency', 'temp_difference']
measures[numerical_cols] = (measures[numerical_cols] - measures[numerical_cols].min()) / (measures[numerical_cols].max() - measures[numerical_cols].min())
```
