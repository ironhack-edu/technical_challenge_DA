#### replacing seperators in first dataframe

```python
# first, replacing ',' with '.'
def replacing_seperators(data1):
    
    data1['distance'] = data1['distance'].str.replace(r',','.')
    data1['temp_inside'] = data1['temp_inside'].str.replace(r',','.')
    data1['consume'] = data1['consume'].str.replace(r',','.')
    
    return data1

data1 = replacing_seperators(data1)
```

#### concateneting both dataframes

```python
data = pd.concat([data1, data2], axis = 0)
data
```

#### converting dtypes

```python
data.dtypes
```

```python
def change_dtypes(data1):
    
    # numbers
    data['distance'] = data['distance'].astype(float) # most likely km or miles
    data['consume'] = data['consume'].astype(float) # moste likely liters
    data['temp_inside'] = data['temp_inside'].astype(float)
    
    # objects: 1 = yes | 0 = no
    data['AC'] = data['AC'].astype(object)
    data['rain'] = data['rain'].astype(object)
    data['sun'] = data['sun'].astype(object)
    
    return data

data = change_dtypes(data)
```

## Cleaning

```python
# columns withour NaNs already clean, focusing on columns with NaNs
```

#### column 'refill liters'

```python
data = data.drop(['refill liters'], axis = 1)
```

#### column 'refill gas'

```python
data = data.drop(['refill gas'], axis = 1)
```

#### column 'specials'

```python
data = data.drop(['specials'], axis = 1)
```

#### column 'temp_inside'

```python
# filling NaNs with average
def clean_temp_inside(data):
    
    # converting object into float, round them to 1 decimal place
    data['temp_inside'] = data['temp_inside'].astype(float)
    
    # filling NaNs with mean value
    data['temp_inside'] = data['temp_inside'].fillna(data['temp_inside'].mean())
    
    # rounding values to 1 decimal place,as thats enough for temperatures
    data['temp_inside'] = data['temp_inside'].apply(lambda x: round(x, 1))
    
    return data

data = clean_temp_inside(data)
```

```python
# renaming AC to ac so its coherent with the rest
data = data.rename(columns={"AC": "ac"})

# resetting index
data = data.reset_index(drop=True)
```

```python
display(data.dtypes)
display(data)
display(data.isnull().sum())
```

#### new consumption per km and then price per km columns
#### based on average prices for E10 (1,38â‚¬) and SP93(1,46â‚¬)

```python
data['consumtion_per_trip'] = (data['consume'] * data['distance']) / 100
data['price_per_trip_e10'] = (data[data['gas_type'] == 'E10']['consumtion_per_trip'] * 1.38)
data['price_per_trip_sp98'] = data[data['gas_type'] == 'SP98']['consumtion_per_trip'] * 1.46
data['price_per_trip_e10'] = data['price_per_trip_e10'].fillna(0)
data['price_per_trip_sp98'] = data['price_per_trip_sp98'].fillna(0)
data['price_per_trip'] = (data['price_per_trip_e10'] + data['price_per_trip_sp98'])
data['consumption_per_km']  = data['consumtion_per_trip'] / data['distance']
data['price_per_km'] = data['price_per_trip'] / data['distance']
data = data.drop(['price_per_trip_e10', 'price_per_trip_sp98'], axis = 1)
```
