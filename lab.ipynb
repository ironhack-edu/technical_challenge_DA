## Data cleaning

```python
# standardize columm names
# lowercase
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
df['sex'].value_counts()
```

```python
def clean_gender(x):
    if x in ['maled', 'M', 'Male','meale','m','male']: # Male gender
        return 'male'
    elif x in ['female','Feamle', 'F']: # Female gender
        return 'female'
    else: # Other
        return 'other'
```

```python
df['sex'] = list(map(clean_gender,df['sex']))
```

```python
df['sex'].value_counts()
```

```python
df['sex'] = df['sex'].map({'male': 1, 'female': 0})
```

```python
df.dtypes
```

```python
# Replace ? with Nan
df = df.replace('?', np.nan)
```

```python
# Cast columns into numerical 
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df.dtypes
```

```python
# Missing Values
df.isnull().sum()
```

Understanding some of the columns with NaNs:

**protime = Prothrombin time** *Higher is worse*


This usually happens because the liver is not making the right amount of blood clotting proteins, so the clotting process takes longer. A high PT usually means that there is serious liver damage or cirrhosis.

**alk_phosphate = ALP Alkaline phosphatase** - *Higher is worse*


It is an enzyme found in the liver and bone and is important for breaking down proteins. Higher-than-normal levels of ALP may indicate liver damage or disease, such as a blocked bile duct, or certain bone diseases.

**albumin = Albumin in blood** - *lower is worse*

 An albumin blood test checks your liver and kidney function. Albumin is protein in your blood plasma. Low albumin levels might be the result of kidney disease, liver disease, inflammation or infections


```python
df.hist(figsize=(20,20))
```

```python
# all low NaNs we can drop as the influence won't be big 
df.dropna(subset=['steroid', 'fatigue', 'malaise', 'anorexia', 'spleen_palpable', 'spiders', 'ascites', 'varices', 'bilirubin', 'sgot'], inplace=True)

# Decision on these to be made 
# liver_big -> This is bad, so to be on the safe side, 1 
df['liver_big'].fillna(1, inplace=True)
# liver_firm -> If tender it's bad, so to be on the safe side, 0
df['liver_firm'].fillna(0, inplace=True)
# alk_phosphate -> slightly skewed to the left, let's use the mean
df['alk_phosphate'].fillna(np.mean(df['alk_phosphate']), inplace=True)
# albumin -> quite well distributed, let's do mean 
df['albumin'].fillna(np.mean(df['albumin']), inplace=True)
# protime -> slightly skewed to the right and with even numbers, let's do median 
df['protime'].fillna(np.median(df['protime']), inplace=True) 
```

```python
df['protime'].fillna(np.mean(df['protime']), inplace=True) 
```
