```python
#import libraries
import pandas as pd
import numpy as np
```

```python
#Importing and reading data
data1 = pd.read_csv('measurements.csv')
data1
```

```python
#Importing and reading data
data2 = pd.read_excel('measurements2.xlsx')
data2
```

### Concatenating

```python
#We observe both of data1 and data2 have the same information so we may concatenate them 
data = pd.concat([data1, data2], axis=0)
data
```

## Details about the data:

Distance (km)
The consume (L/100km)
The average speed (km/h)
The temperature inside (°C)
The temperature outside (°C)
AC (on=1/off=0)
The sun (sunny=1)
Fuel price per gas type
Calculate fuel consume per km
Calculate cost per trip
Calculate cost per kilometer traveled 

```python
data.columns = data.columns.str.replace(' ', '_')
data
```

```python
#Checking for NaN values 
data.isna().sum()
```

```python
#Cheking the type of each column
data.dtypes
```

```python
#Replacing in columns the "," into "." as we need to convert them to numeric
data['temp_inside'] = data['temp_inside'].str.replace(',', '.').fillna(0).astype(float)
data['distance'] = data['distance'].str.replace(',', '.').fillna(0).astype(float)
data['consume'] = data['consume'].str.replace(',', '.').fillna(0).astype(float)
data['refill_liters'] = data['refill_liters'].str.replace(',', '.').fillna(0).astype(float)
```

```python
#Checking for duplicates
data.duplicated().sum()
```

```python
#Dropping duplicates
data = data.drop_duplicates()
```

```python
#Checking for NaN values again
data.isna().sum()
```

```python
data.isna().sum()
```

### Selecting numerical and categorical columns

```python
#Creating only numerical data
numerical = data.select_dtypes(np.number)
numerical.head()
```

```python
#Checking the values in the column as to replace the NaN values
numerical.isna().sum()
```

```python
#Selecting the categorical columns
categorical = data.select_dtypes(object)
categorical.head()
```

```python
categorical.isna().sum()
```

```python
#Checking the values in the column as to replace the NaN values
categorical['specials'].value_counts()
```

```python
#Decided to replace the NaN values with another value
categorical['specials'] = categorical['specials'].fillna('other')
```

```python
#Checking the values in the column as to replace the NaN values
categorical['refill_gas'].value_counts()
```

```python
#Filling the NaN values at column refill gas with what exists at column gas type
def gas_fill(refill_gas, gas_type):
    if pd.isna(refill_gas):
        return gas_type
    else:
        return refill_gas

categorical['refill_gas'] = categorical.apply(lambda row: gas_fill(row['refill_gas'], row['gas_type']), axis=1)
categorical['refill_gas'].isna().sum()
```

```python
categorical.isna().sum()
```

```python
categorical.info
```

### Concatenating numerical and categorical 

```python
numerical.reset_index(drop=True, inplace=True)
categorical.reset_index(drop=True, inplace=True)

data_final = pd.concat([numerical, categorical], axis=1)
data_final
```

```python
#Add new column with the fuel price
data_final['fuel_price_liter'] = data_final['gas_type'].apply(lambda x: 1.38 if x == 'E10' else 1.46)
```

```python
#Calculate fuel consume per km
data_final['fuel_consume_km'] = (data_final['consume']*data_final['distance'])/100
#Calculate cost per trip
data_final['trip_cost'] = data_final['fuel_consume_km']*data_final['fuel_consume_km']
#Calculate cost per kilometer traveled 
data_final['cost_km'] = data_final['trip_cost']/data_final['distance']
data_final.head()
```

```python
data_final['cost_km'] = data_final['cost_km'].fillna(data_final['cost_km'].mean())
data_final.isna().sum()
```

```python
data_final.to_csv('datacleanedup.csv', index=False)
```
