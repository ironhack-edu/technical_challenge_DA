# Hypothesis testing

```python
import scipy.stats as st
```

```python
# H0 = consume E10 = consume SP98
# H1 = consume E10 != consume SP98
```

```python
SP98 = new_df2[new_df2['gas_type']==1]
E10 = new_df2[new_df2['gas_type']==0]
```

```python
st.ttest_ind(SP98['consume'], E10['consume'], equal_var=False)
```

Having p value = 0.67 we don't have enough evidence to say that the consumption of gas is influenced by its type

```python
# H0 = AC on consume = AC off consume
# H1 = AC on consume != AC off consume
```

```python
on = new_df2[new_df2['AC']==1]
off = new_df2[new_df2['AC']==0]
```

```python
st.ttest_ind(on['consume'], off['consume'], equal_var=False)
```

Having p value = 0.23 we don't have enough evidence to say that the consumption of gas is influenced by the use of AC

## Cleaning function

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
def get_ready(X,y):
    # * Split the data into a training set and a test set.
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
    print(X_train.shape)
    print(X_test.shape)
    print(y_train.shape)
    print(y_test.shape)
     
    
    # * Scale the features either by using normalizer or a standard scaler.
        
    #fit and transform scaling numericals
    minmax = MinMaxScaler().fit(X_train) 
    
    X_train_norm = minmax.transform(X_train)
    X_test_norm = minmax.transform(X_test)
    
    # convert to dataframe again
    X_train_norm = pd.DataFrame(X_train_norm,columns=X_train.columns).reset_index(drop=True) 
    X_test_norm = pd.DataFrame(X_test_norm,columns=X_test.columns).reset_index(drop=True) 

    return X_train_norm, X_test_norm, y_train, y_test
```

## Choosing model

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
def models_automation(models, X_tr, y_tr, X_te, y_te):
    for model in models:
        model.fit(X_tr, y_tr)
        print(f"{model.__class__.__name__}: Train -> {model.score(X_tr, y_tr)}, Test -> {model.score(X_te, y_te)}")
    return
```

# Looking into predicting consume

```python
y = new_df2['consume']
X = new_df2.drop(['consume'],axis=1)
```

```python
X_train, X_test, y_train, y_test = get_ready(X,y)
```

```python
models = [LinearRegression(),SGDRegressor(), DecisionTreeRegressor(),KNeighborsRegressor(),RandomForestRegressor()]

models_automation(models, X_train,y_train,X_test,y_test)
```

# Decision Tree Regressor

```python
# why dont I repeat the process but choose the max_depth of the tree
max_depth = range(1,30)
test = []
train = []

for depth in max_depth:
    model = DecisionTreeRegressor(max_depth=depth, random_state=0)
    model.fit(X_train, y_train)
    test.append(model.score(X_test,y_test))
    train.append(model.score(X_train,y_train))
    
    
    
# we quickly see overfitting properties
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot( train, label="training accuracy")
plt.plot( test, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_depth-1")
plt.legend()
```

```python
model = DecisionTreeRegressor(max_depth=8)
model.fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
model.score(X_test, y_test)
```

```python
cross_val_score(model, X_train, y_train, cv=10)
```

```python
from matplotlib.pyplot import figure

figure(figsize=(8, 12), dpi=80)

def plot_feature_importances(model, namelist):
    n_features = len(model.feature_importances_)
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), namelist)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")

plot_feature_importances(model, X_train.columns)
```
