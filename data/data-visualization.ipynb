```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ms = pd.read_csv('measurements-cleaned.csv')
```

```python
gt = []
for e in ms.gas_type:
    if e == 'E10':
        gt.append(10)
    else:
        gt.append(98)
```

```python
ms['gas_type_num'] = gt
```

```python
plt.figure(figsize=(8, 8))

sns.set(style='darkgrid')

mask=np.triu(np.ones_like(ms.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(ms.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

Although there is not a high correlation in the data, the highest we can find is between speed and distance, which makes sense taking into consideration that if you drive far away you might be taking high speed roads.

We should also highlight the fact that I wanted to check the possibility of existing a correlation between gas type and consume, that is why I created a new columns with numeric data representing gas type, but this columns does not reveal anything.

```python
plt.figure(figsize=(8, 8))

sns.boxplot(data = ms);
```

The above boxplot shows that the distance column is the one with most outliers and the second one is speed. Taking into consideration the previous correlation figure, this makes sense.
