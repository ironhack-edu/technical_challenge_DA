```python
# There are duplicate values like AC, ac. So grouping them under single value

def clean_specials(value):
    
    if value == "ac":
        return "AC"
    elif (value == "AC sun") | (value == "sun ac"):
        return "AC Sun"
    elif (value == "ac rain"):
        return "AC rain"
    elif value == "half rain half sun":
        return "rain"
    elif value == "AC snow":
        
        # Combining AC snow under snow as well because there is only one value in it. And it might give error during machine learning
        return "snow"
    else:
        return value

data.specials = data.specials.apply(clean_specials)
```

# Hypothesis Testing

```python
Sp98 = data[data.gas_type == "SP98"].reset_index(drop = True)
E10 = data[data.gas_type == "E10"].reset_index(drop = True)
```

```python
Sp98.head()
```

```python
E10.head()
```

```python
# H0: There is no difference in fuel consumption for E10 and SP98.
# H1: There is difference in fuel consumption for E10 and SP98.

st.ttest_ind(Sp98["consume"], E10["consume"], equal_var=False)
```

```python
print(Sp98.consume.mean())
print(E10.consume.mean())
```

```python
# Conclusion: As p-val is greater than 0.05. I accept the null hypothesis. 

# Although the mean value isn't same but they are close enough.
```

```python
features = ["distance", "consume", "speed", "temp_inside", "temp_outside", "refill_liters", "consume_liter"]

def t_test_gas_type(s1, s2, features = features):
    
    results = {}
        
    for feature in features:
        
        result = st.ttest_ind(s1[feature], s2[feature], equal_var=False)
        
        # Using only p-value in dictionary. 
        results[feature] = result[1]

    return results
```

```python
# H0: The mean of SP98 and E10 features are same.
# H1: The mean of SP98 and E10 features are not same.

print(t_test_gas_type(Sp98, E10))
```
